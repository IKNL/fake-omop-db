```python
important_sheets = [
 'Patient',
 'PatientFollowUp',
 'HospitalData',
 'HospitalPatientRecords',
 'CancerEpisode',
 'Diagnosis',
 'ClinicalStage',
 'PathologicalStage',
 'EpisodeEvent',
 'Disease extent',
 'GeneticTestExpression',
 'Surgery',
 'SystemicTreatment',
 'Radiotherapy',
 'RegionalDeepHyperthemia',
 'IsolatedLimbPerfusion',
 'DrugsForTreatments',
 'OverallTreatmentResponse',
 'AdverseEvent',
]
```

```python
# substring = 'CONDITION'
# substring = 'PROCEDURE'
# substring = 'MEASUREMENT'
# substring = 'OBSERVATION'
substring = 'EPISODE'

# Initialize an empty DataFrame to store the results
combined_df = pd.DataFrame()

# Loop through all the sheet names
for sheet_name in important_sheets:
    # Read the current sheet into a DataFrame
    df = pd.read_excel(file, sheet_name=sheet_name)
    
    # Filter rows that contain 'CONDITION' in the 'omop' column
    condition_rows = df[df['Value in OMOP'].str.contains(substring, na=False)]
    condition_rows.insert(0, 'sheet_name', sheet_name)
    
    # Append the filtered rows to the combined DataFrame
    combined_df = pd.concat([combined_df, condition_rows], ignore_index=True)

# combined_df.head()
combined_df[['sheet_name', 'Variable Name (EURACAN file)','Value in OMOP', 'ExpectedValue', 'Variable as modifier', 'Vocabulary']]
```
